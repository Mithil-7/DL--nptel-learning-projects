### SLP SINGLE LAYER PERCEPTOR

```python
import numpy as nd
```

```python
x=nd.array([1.0,2.0])
y=1
w=nd.array([0.4,-0.6])
bias=0.2
eta=0.1
```

```python
for epoch in range(1,4):
  print(f"\nEpoch {epoch}:")
  net_input=nd.dot(w,x)*eta + bias
  y_pred=1 if net_input>0 else 0
  error=y-y_pred

  print(f"  Net input : {net_input:.2f}")
  print(f"  Predicted : {y_pred}")
  print(f"  Error     : {error}")


  w=error*eta*x + w
  bias= error*eta +bias

  print(f"Updated weight : {w}")
  print(f"Updated bias   : {bias:.2f}")
```
